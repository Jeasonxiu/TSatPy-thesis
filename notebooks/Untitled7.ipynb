```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import rcParams
from mpl_toolkits.mplot3d import Axes3D
from TSatPy import State
from TSatPy.Clock import Metronome

rcParams['text.usetex']=True
rcParams['text.latex.unicode']=True


def gen_frames(clock, w, ics):

    # Setup for 1/4 turn each sec about the +z axis
    systems = []
    
    for ic in ics:
        qd = State.QuaternionDynamics(ic, clock)
        systems.append(qd)
    
    def frames():
        for i in range(1000):
            clock.set_speed(np.sin(i/30) * 0.4 + 0.5)
            yield [sys.propagate(w) for sys in systems]
    return frames

        
def func(states, clock, models, time_text):
    time_text.set_text(r"""clock speed = $%g$
$\mathbf{q} = %s$""" % (clock.speed, states[0].latex().replace('boldsymbol', 'mathbf')))
    
    for q, model in zip(states, models):
        for name, sdata in model.series.items():
            p = q.rotate_points(sdata['points'])
            x, y, z = p.T        
            sdata['plot'].set_data(np.asarray(x).reshape(-1), np.asarray(y).reshape(-1))
            sdata['plot'].set_3d_properties(np.asarray(z).reshape(-1))


class TSatModel():
    def __init__(self, ax, color='b'):
        count = 2
        radius = 0.2
        sdp_len = 0.6
        zeros = np.empty(count + 1)
        zeros.fill(0)

        self.series = {
            '+x sdp': {
                'points': np.mat([np.arange(count + 1) * (sdp_len / float(count)) + radius,
                    zeros.copy(), 
                    zeros.copy()]).T,
                'plot': ax.plot([], [], [], color=color, ls='-', linewidth=2, ms=10)[0],
            },
            '-x sdp': {
                'points': np.mat([-(np.arange(count + 1) * (sdp_len / float(count)) + radius),
                    zeros.copy(), 
                    zeros.copy()]).T,
                'plot': ax.plot([], [], [], color=color, ls='-', linewidth=2, ms=10)[0],
            },
            '+y sdp': {
                'points': np.mat([zeros.copy(),
                    np.arange(count + 1) * (sdp_len / float(count)) + radius,
                    zeros.copy()]).T,
                'plot': ax.plot([], [], [], color=color, ls='-', linewidth=2, ms=10)[0],
            },
            '-y sdp': {
                'points': np.mat([zeros.copy(),
                    -(np.arange(count + 1) * (sdp_len / float(count)) + radius),
                    zeros.copy()]).T,
                'plot': ax.plot([], [], [], color=color, ls='-', linewidth=2, ms=10)[0],
            },
            '+z adp': {
                'points': np.mat([zeros.copy(),
                    zeros.copy(),
                    np.arange(count + 1) * (sdp_len / float(count)) + radius]).T,
                'plot': ax.plot([], [], [], color=color, ls='-', linewidth=2, ms=10)[0],
            },
            '-z adp': {
                'points': np.mat([zeros.copy(),
                    zeros.copy(),
                    -(np.arange(count + 1) * (sdp_len / float(count)) + radius)]).T,
                'plot': ax.plot([], [], [], color=color, ls='-', linewidth=2, ms=10)[0],
            },
        }

def main():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d', aspect='equal')

    ax.set_ylim(-1, 1)
    ax.set_xlim(-1, 1)
    ax.set_zlim(-1, 1)

    time_text = ax.text(0.05, 0.9, 0, '', transform=ax.transAxes)
    
    w = State.BodyRate([0, 0, np.pi/8])

    ics = [
        State.Quaternion([0,0.1,1], radians=np.pi/4),
        State.Identity()
    ]

    clock = Metronome()
    clock.set_speed(0.5)

    frames = gen_frames(clock, w, ics)
    refresh_rate = 0.3
    
    models = []
    models.append(TSatModel(ax))
#     models.append(TSatModel(ax, 'g'))
    
    kwargs = {
        'fig': fig,
        'func': func,
        'frames': frames,
        'blit': False, 
        'interval': float(refresh_rate) * 1000,
        'fargs': [clock, models, time_text],
    }
    ani = animation.FuncAnimation(**kwargs)
    
    plt.show()

main()
```
