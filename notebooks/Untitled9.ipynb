```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import rcParams
from mpl_toolkits.mplot3d import Axes3D
from TSatPy import State
from TSatPy.Clock import Metronome

rcParams['text.usetex']=True
rcParams['text.latex.unicode']=True


def gen_frames(clock, sim_time):
    end_time = clock.tick() + sim_time
    def frames():
        frame_count = 0
        while clock.tick() < end_time:
            yield frame_count
            frame_count += 1
    return frames


class TSatModel():
    def __init__(self, ax, color='b'):
        self.ax = ax
        self.color = color
        self.series = {}
        self.radius = 0.2
        self.add_booms()
    
    def add_booms(self):
        count = 2
        sdp_len = 0.6
        zeros = np.empty(count + 1)
        zeros.fill(0)

        # sdp's
        self.series['+x sdp'] = {
            'points': np.mat([np.arange(count + 1) * (sdp_len / float(count)) + self.radius,
                zeros.copy(), 
                zeros.copy()]).T,
            'plot': self.ax.plot([], [], [], color=self.color, ls='-', linewidth=2, ms=10)[0],
        }
        self.series['-x sdp'] = {
            'points': np.mat([-(np.arange(count + 1) * (sdp_len / float(count)) + self.radius),
                zeros.copy(), 
                zeros.copy()]).T,
            'plot': self.ax.plot([], [], [], color=self.color, ls='-', linewidth=2, ms=10)[0],
        }
        self.series['+y sdp'] = {
            'points': np.mat([zeros.copy(),
                np.arange(count + 1) * (sdp_len / float(count)) + self.radius,
                zeros.copy()]).T,
            'plot': self.ax.plot([], [], [], color=self.color, ls='-', linewidth=2, ms=10)[0],
        }
        self.series['-y sdp'] = {
            'points': np.mat([zeros.copy(),
                -(np.arange(count + 1) * (sdp_len / float(count)) + self.radius),
                zeros.copy()]).T,
            'plot': self.ax.plot([], [], [], color=self.color, ls='-', linewidth=2, ms=10)[0],
        }
        self.series['+z adp'] = {
            'points': np.mat([zeros.copy(),
                zeros.copy(),
                np.arange(count + 1) * (sdp_len / float(count)) + self.radius]).T,
            'plot': self.ax.plot([], [], [], color=self.color, ls='-', linewidth=2, ms=10)[0],
        }
        self.series['-z adp'] = {
            'points': np.mat([zeros.copy(),
                zeros.copy(),
                -(np.arange(count + 1) * (sdp_len / float(count)) + self.radius)]).T,
            'plot': self.ax.plot([], [], [], color=self.color, ls='-', linewidth=2, ms=10)[0],
        }

def func(frame, clock, models, plants, time_text):
    time_text.set_text(r"clock speed = $%g$" % (clock.speed))
    
#     , states[0].latex().replace('boldsymbol', 'mathbf')

    for name in models.keys():
        plants[name].propagate()
        x = plants[name].x
        for name, sdata in model.series.items():
            p = x.q.rotate_points(sdata['points'])
#             x, y, z = p.T        
            sdata['plot'].set_data(np.asarray(p[:,0].T).reshape(-1), np.asarray(p[:,1].T).reshape(-1))
            sdata['plot'].set_3d_properties(np.asarray(p[:,2].T).reshape(-1))


def main():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d', aspect='equal')

    ax.set_ylim(-1, 1)
    ax.set_xlim(-1, 1)
    ax.set_zlim(-1, 1)

    time_text = ax.text(0.05, 0.9, 0, '', transform=ax.transAxes)

    clock = Metronome()
    clock.set_speed(1)

    refresh_rate = 0.3

    x = State.State(
        State.Quaternion([0, 0, 1], radians=np.pi/2),
        State.BodyRate([0, 0, 0.1]),
    )
    I = [[4, 0, 0], [0, 4, 0], [0, 0, 4]]

    plants = {
        'truth': State.Plant(I, x, clock)
    }

    frames = gen_frames(clock, sim_time=10)
    
    models = {
        'truth': TSatModel(ax)
    }
#     models.append(TSatModel(ax, 'g'))
    
    kwargs = {
        'fig': fig,
        'func': func,
        'frames': frames,
        'blit': False, 
        'interval': float(refresh_rate) * 1000,
        'fargs': [clock, models, plants, time_text],
    }
    ani = animation.FuncAnimation(**kwargs)
    
    plt.show()

main()
```
